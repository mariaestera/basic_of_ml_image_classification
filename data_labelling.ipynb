```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
IMAGE_DIR = "data//images"

image_paths = [
    os.path.join(IMAGE_DIR, fname)
    for fname in os.listdir(IMAGE_DIR)
    if fname.lower().endswith(".png")
]

print(f"{len(image_paths)} images was founded")
```

```python
label_images = [fname
    for fname in os.listdir("data\\images_labelled")
    if fname.lower().endswith(".png")
]

for im_path in image_paths:
    name = im_path.split("\\")[-1]
    if any(name in x for x in label_images):
        continue
    
    img = cv2.imread(im_path, cv2.IMREAD_GRAYSCALE)

    cv2.imshow("Obraz", img)
    cv2.waitKey(1)  
    digit = input("Wpisz liczbÄ™: ")
    
    new_name = f"data\\images_labelled\\{digit}-{name}.png"
    cv2.imwrite(new_name, img)

    cv2.destroyAllWindows()
labels = [fname.split("-")[0]
    for fname in os.listdir("data\\images_labelled")
    if fname.lower().endswith(".png")
]
```

```python
labels = [[int(y) for y in list(x)] for x in labels]
numbers = [item for sublist in labels for item in sublist]
```

```python
sns.histplot([len(x) for x in labels], bins =8)
```

```python
numbers_arr = np.array(numbers)
counts = np.bincount(numbers_arr, minlength=10)

for digit, c in enumerate(counts):
    print(f"Digit {digit}: {c} times")
```
